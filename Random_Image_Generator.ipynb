```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Generate random pixel data (e.g., grayscale or RGB)
width, height = 512, 512 # Changed dimensions
random_pixels = np.random.randint(0, 256, (height, width, 3), dtype=np.uint8) # For RGB

# Create an image from the pixel data
random_image = Image.fromarray(random_pixels)

# Display the image
plt.imshow(random_image)
plt.axis('off') # Hide axes
plt.title("Random Color Image")
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Generate random pixel data (e.g., grayscale or RGB)
width, height = 512, 512
random_pixels = np.random.randint(0, 256, (height, width), dtype=np.uint8) # For grayscale

# Create an image from the pixel data
random_image = Image.fromarray(random_pixels)

# Display the image
plt.imshow(random_image, cmap='gray')
plt.axis('off') # Hide axes
plt.title("Random Grayscale Image")
plt.show()
```
